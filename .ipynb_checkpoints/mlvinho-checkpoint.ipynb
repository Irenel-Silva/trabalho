```python
# conjunto de dados de análise de vinho
import pandas as pd
vinhods=pd.read_csv('C:/Users/Irenel/anaconda3/projeto/analisedados/vinhoanalise/wine_dataset.csv')
```

```python
vinhods.head(6)
```

```python
vinhods.shape
```

```python
# filtrar a coluna style para alterar a string para numero a ser estudado pelo nosso algoritmo machine learning

vinhods['style']=vinhods['style'].replace('red',0)
vinhods['style']=vinhods['style'].replace('white',1)
vinhods.head()
```

```python
# separando as variavéis entre preditoras e variavéis alvo
y= vinhods['style']
x= vinhods.drop('style', axis=1)
```

```python
# preparando os dados de testes e os dados de treino

from sklearn.model_selection import train_test_split
# criando o conjunto de dados de teste e de treino:
x_treino, x_teste, y_treino, y_teste= train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
#criação do modelo:
modelo=ExtraTreesClassifier(n_estimators=100)
modelo.fit(x_treino,y_treino)

# imprimindo resultados:
resultado= modelo.score(x_teste, y_teste)
print('Acurácia', resultado)
```
